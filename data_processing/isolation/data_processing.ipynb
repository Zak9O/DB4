```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.array(pd.read_csv("data.csv"))
# Skip the first part of the data to make it match the other data
data = data[140:] 
data[:, 0] -= data[0,0]
data
```

```python
data_without = np.array(pd.read_csv("data_without.csv"))
data_without
```

```python
fig, ax = plt.subplots()
ax.plot(data_without[:,0], data_without[:,1])
ax.plot(data[:,0], data[:,1])
ax.legend(["without","with"])
ax.set_xlabel("Time (s)")
ax.set_ylabel("Temperature (C)")
ax.set_title("Temperature decay")
plt.show()
```
